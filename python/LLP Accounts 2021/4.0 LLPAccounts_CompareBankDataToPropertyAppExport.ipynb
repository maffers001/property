### Split rent dates - may need to move this somewhere else
- this is used to identify rents that cross start/end periods of financial year
- probably should be put in journals generation

```python
import re
from datetime import datetime

start_period=datetime.strptime('06/04/2021', "%d/%m/%Y").date()
end_period=datetime.strptime('05/04/2022', "%d/%m/%Y").date()

for index, row in df6[(df6.Category=='RentalIncome')].iterrows():
    #print('>>>>>>>>>>>>>>>>>>>>>>>>>')
    search=re.search("^Rent for period ([0-9]{2}\/[0-9]{2}\/[0-9]{4})\s*-\s*([0-9]{2}\/[0-9]{2}\/[0-9]{4}) from.*$", row.Description)
    #print(search.group(1))
    #print(search.group(2))
    start_date_str=search.group(1)
    if(start_date_str=='30/02/2022'): start_date_str='01/03/2022' # Correct dates
    start_date=datetime.strptime(start_date_str, "%d/%m/%Y").date()
    end_date_str=search.group(2)
    if(end_date_str=='29/02/2022'): end_date_str='01/03/2022'  # Correct dates
    if(end_date_str=='30/02/2022'): end_date_str='01/03/2022'  # Correct dates
    end_date=datetime.strptime(end_date_str, "%d/%m/%Y").date()
    if((start_date<start_period) and (end_date>start_period) ):
        print(row)
    if((start_date<end_period) and (end_date>end_period) ):
        print(row)

        
        
# Identify rent period records
#df6[(df6.Category=='RentalIncome')&(df6.Description.str.contains("Rent for period ([0-9]{2}\/[0-9]{2}\/[0-9]{4}) - ([0-9]{2}\/[0-9]{2}\/[0-9]{4})"))]    
```
