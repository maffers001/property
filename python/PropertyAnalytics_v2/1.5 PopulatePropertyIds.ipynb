```python
def check_existing_propertyids(dfB):
    for index, row in dfB[(~((pd.isnull(dfB.Property))|(dfB.Property=='')))].iterrows():                
        if(row.Property not in propertyidmap):
            print('Not found in property id: ' + row.Property)
            dfB.at[index,'Property']=''
    return dfB

def allocate_codes_for_mortgages(dfB):
    dfB=dfB.reset_index()
    # Allocate property IDs to Bank payments from Beals
    for index, row in dfB[(dfB.Subcategory.str.contains('DIRECTDEBIT|OTH|PAYMENT|Direct Debit|Other|Bill Payment',case=False,regex=True))&((pd.isnull(dfB.Property))|(dfB.Property==''))].iterrows():                
        if(isinstance(row.Memo, str)):
            for key in mortgage_map:
                if(re.match(key,row.Memo)):
                    dfB.at[index,'Property']=mortgage_map[key]
    dfB.set_index(dfB.Date,inplace=True)
    dfB.drop('Date', axis=1, inplace=True)
    return dfB
```

### Populate property ids for files between start/end dates

- input files are 'MMMYYYY.csv'
- allocates property ids to mortgages and property expenses
- the output files are named 'MMMYYYY_coded.csv'

```python
from dateutil.rrule import rrule, MONTHLY

dfDD= pd.DataFrame()

dates = [dt for dt in rrule(MONTHLY, dtstart=start, until=end)]
for date in dates:
    dateStr= date.strftime("%b").upper() + date.strftime("%Y")
    input_file=generated_path + dateStr + '.csv'
    print('Processing: '+ input_file)
    df=pd.read_csv(input_file, index_col=0, parse_dates=True, dayfirst=True)
    if 'Property' not in df.columns:
        df['Property']=''
    else:
        df['Property'] = df['Property'].astype(str)
        df.loc[(df.Property=='nan'),'Property']=''
        
    df=check_existing_propertyids(df)
    df=allocate_codes_for_mortgages(df)
    df=allocate_codes_for_rents_and_expenses(df)
    file=backup_file(generated_path + dateStr + '_coded.csv')
    df.to_csv(file)   
    dfTemp=df.loc[((df.Subcategory.str.match('.*DIRECT[ ]?DEBIT.*',case=False)==True)|(df.Memo.str.match('^BEALS.*$',case=False)==True))]
    dfDD=pd.concat([dfDD,dfTemp])
    
# Use this to check propertyIds for Mortgages and Beals payments in one place    
dfDD.to_csv('DDCheck.csv')
```
